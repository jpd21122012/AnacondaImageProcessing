```python
#So to check if we have opencv installed we need to type
import cv2
import imgaug as ia
```

```python
# and now we can use this library to work with our videos or images.........

#Let's find a good image to work with that...

#So to read an image locally or on web we need the following method

img = cv2.imread("helena.jpg")
ia.imshow(img)
#Alright it appears that now we cannot read like that images from internet so I'm going to download it...

#cv2.waitKey(25)
```

```python
#So to print the dimentions of this image we perform the .shape like this...

print(img.shape[0:3])
```

```python
# We have width height and the channels, we are reading this image with RGB System

#To rotate the image we have to call the rotationMatrix method...

#I missed the width and height...

height,width = img.shape[0:2]

#So in this method the parameters are: (center,angle,scale), so we'll perform a rotation of 90 and the half of scale
rotationMatrix = cv2.getRotationMatrix2D((width/2, height/2),90, .5)

rotatedImage = cv2.warpAffine(img,rotationMatrix,(width,height))

#And finally we have.........

ia.imshow(rotatedImage)

```

```python
# The next step is to Crop the image

startRow = int(height*.15)
startCol = int(width*.15)
endRow = int(height*.85)
endCol = int(width*.85)
```

```python
#Assign the sizes on an array

croppedImage = img[startRow:endRow,startCol:endCol]
```

```python
#So now we can visualize the image..

ia.imshow(img)
ia.imshow(croppedImage)
```

```python
# To resize the image we use resize() method..
newImg = cv2.resize(img,(0,0),fx=.75,fy=.75)

ia.imshow(newImg)

```

```python
#Let's put manually the values of rows and columns (height and width)

newImg = cv2.resize(img,(750,750))

ia.imshow(newImg)

```

```python
# to adjust the contrast we can use the AddWeighted method like this..
#import array handler

import numpy as np
contrast_img = cv2.addWeighted(img,2.5,np.zeros(img.shape,img.dtype),0,0)

ia.imshow(contrast_img)
```

```python
# Make a Blurred Image using Gaussian Methods...

blur_image = cv2.GaussianBlur(img, (7,7),0)

ia.imshow(blur_image)
```

```python
#To make a Median Blur use this...
medianBlur_image = cv2.medianBlur(img,5)

ia.imshow(medianBlur_image)
```

```python
# Now we can use an Edge Filter..... 
edge_img = cv2.Canny(img,100,200)
ia.imshow(edge_img)
```

```python
# We learned to process the image with colors, but what if we want the grayscale Image.....

gray_img = cv2.imread("helena.jpg",0)

ia.imshow(gray_img)
```
